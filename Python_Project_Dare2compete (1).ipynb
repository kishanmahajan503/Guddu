We will create Linear Regression Model 

```python
Linear_Regression = df_nonan[['fuel-type', 'aspiration', 'drive-wheels', 'wheel-base', 'length', 'width', 'curb-weight', 'engine-type',
       'num-of-cylinders', 'engine-size', 'horsepower', 'price', 'highway-mpg']]
Linear_Regression.head()
```

Train-Test Split

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(Linear_Regression, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import LabelEncoder

number = LabelEncoder()
df_train["fuel-type"] = number.fit_transform(df_train["fuel-type"].astype('str'))
df_test["fuel-type"] = number.fit_transform(df_test["fuel-type"].astype('str'))

df_train["drive-wheels"] = number.fit_transform(df_train["drive-wheels"].astype('str'))
df_test["drive-wheels"] = number.fit_transform(df_test["drive-wheels"].astype('str'))

df_train["aspiration"] = number.fit_transform(df_train["aspiration"].astype('str'))
df_test["aspiration"] = number.fit_transform(df_test["aspiration"].astype('str'))

df_train["engine-type"] = number.fit_transform(df_train["engine-type"].astype('str'))
df_test["engine-type"] = number.fit_transform(df_test["engine-type"].astype('str'))

df_train["num-of-cylinders"] = number.fit_transform(df_train["num-of-cylinders"].astype('str'))
df_test["num-of-cylinders"] = number.fit_transform(df_test["num-of-cylinders"].astype('str'))

df_train.head()
```
